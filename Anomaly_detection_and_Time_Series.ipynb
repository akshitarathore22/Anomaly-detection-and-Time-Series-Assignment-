# Anomaly detection and Time Series

Q1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

A1. **Anomaly Detection** is a data analysis technique used to identify unusual patterns or observations in a dataset that do not conform to expected behavior. These anomalies may indicate important events such as fraud, system failures, errors, or rare occurrences. It is widely used in domains like finance, healthcare, cybersecurity, and manufacturing to detect irregularities and take corrective actions.

There are three main types of anomalies. A **point anomaly** (or global anomaly) occurs when a single data point is significantly different from the rest of the data. For example, if most daily transactions of a customer are between ₹500 and ₹2000, but suddenly a transaction of ₹1,00,000 occurs, it is considered a point anomaly. A **contextual anomaly** (or conditional anomaly) depends on the context such as time, location, or season. A data point may be normal in one situation but abnormal in another. For instance, a temperature of 25°C is normal during summer but unusual during winter, making it a contextual anomaly. A **collective anomaly** occurs when a group of data points together behaves abnormally, even though individual points may appear normal. For example, a series of small transactions happening rapidly within a short time may indicate fraudulent activity, even if each transaction alone does not seem suspicious.

Thus, anomaly detection helps in identifying different types of unusual patterns—whether in individual data points, specific contexts, or groups of observations—making it a crucial tool in data science.


Q2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.

A2. Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are popular anomaly detection techniques, but they differ in their approach and suitable use cases.

Isolation Forest is a tree-based algorithm that works on the idea of isolating anomalies rather than profiling normal data. It randomly selects features and splits values to create decision trees. Since anomalies are rare and different, they get isolated quickly in fewer splits, resulting in shorter path lengths in the trees. This method is efficient for large and high-dimensional datasets and works well when anomalies are few and distinct. It is commonly used in fraud detection, network security, and large-scale data analysis.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm that groups data points based on density. It identifies dense regions as clusters and labels points that do not belong to any cluster as noise (anomalies). DBSCAN is effective for detecting anomalies in datasets where clusters of arbitrary shapes exist and where noise points are clearly separated. It is useful in spatial data analysis, image processing, and geographical data, but it may struggle when data has varying densities or in high-dimensional spaces.

Local Outlier Factor (LOF) is a density-based method that detects anomalies by comparing the local density of a data point with that of its neighbors. If a point has significantly lower density than its surrounding points, it is considered an outlier. LOF is particularly useful for detecting local anomalies in datasets where normal data points may have different densities. It works well in applications like intrusion detection and fraud detection where local deviations are important, but it can be computationally expensive for large datasets.

In summary, Isolation Forest focuses on isolating anomalies and is suitable for large datasets, DBSCAN identifies anomalies as noise based on clustering and is good for spatial data, and LOF detects anomalies based on local density differences, making it effective for identifying subtle, context-based outliers.

Q3. What are the key components of a Time Series? Explain each with one
example.

A3. A Time Series is a sequence of data points recorded at regular intervals over time, such as daily sales, monthly temperature, or yearly population. Time series data typically consists of several key components that help in understanding patterns and making forecasts.

The first component is the Trend, which represents the long-term movement or direction in the data over a period of time. It shows whether the values are generally increasing, decreasing, or remaining constant. For example, the steady increase in online shopping sales over several years indicates an upward trend.

The second component is Seasonality, which refers to regular and repeating patterns that occur at fixed intervals due to seasonal factors like weather, festivals, or holidays. For instance, ice cream sales usually increase during summer every year, showing a seasonal pattern.

The third component is the Cyclical Component, which represents fluctuations that occur over a longer period and are not fixed like seasonality. These are often influenced by economic or business cycles. For example, during an economic boom, company profits may rise, while during a recession, they may fall, forming a cycle.

The fourth component is Irregular or Random Variation, which includes unpredictable and random fluctuations in the data caused by unexpected events. These cannot be easily forecasted. For example, a sudden drop in sales due to a natural disaster or pandemic is an irregular component.

Thus, time series analysis involves understanding these components—trend, seasonality, cyclical patterns, and irregular variations—to better interpret data and make accurate predictions.

Q4. Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

A4. A stationary time series is one whose statistical properties remain constant over time. This means that the mean, variance, and autocorrelation structure do not change as time progresses. In simple terms, the data does not show trends, seasonal effects, or changing variability. Stationarity is important because many time series models, such as ARIMA, assume that the data is stationary for accurate forecasting.

To test whether a time series is stationary, both visual and statistical methods are used. Visually, we can plot the data and check for trends, seasonality, or changing variance. A stationary series will appear relatively flat with constant fluctuations around a fixed mean. Statistically, tests like the Augmented Dickey-Fuller (ADF) test and KPSS test are commonly used. In the ADF test, if the p-value is less than a chosen significance level (usually 0.05), the series is considered stationary. On the other hand, the KPSS test checks for stationarity directly, where a small p-value indicates non-stationarity.

If a series is non-stationary, it can be transformed into a stationary one using several techniques. One common method is differencing, where we subtract the previous value from the current value to remove trends. Another method is applying logarithmic or square root transformations to stabilize variance when data shows increasing fluctuations. Detrending can also be used to remove long-term trends, and seasonal differencing helps eliminate seasonal patterns. For example, if monthly sales show a steady upward trend, taking the difference between consecutive months can help make the series stationary.

Thus, stationarity ensures stable statistical properties in time series data, and testing and transforming non-stationary data is a crucial step before applying forecasting models.

Q6. Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
data = get_rdataset("AirPassengers").data

# Convert to time series
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)

# Rename column
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Plot original series
plt.figure(figsize=(10,5))
plt.plot(data['Passengers'], label='Original Series')
plt.title('AirPassengers Time Series')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.show()

# Decomposition
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative', period=12)

# Plot components
decomposition.plot()
plt.show()
```

Q7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# --- Step 1: Create sample dataset (similar to taxi fare data) ---
np.random.seed(42)

# Normal data (fare vs distance)
fare = np.random.normal(200, 50, 300)
distance = np.random.normal(5, 2, 300)

# Add anomalies
fare_outliers = [500, 600, 50]
distance_outliers = [20, 25, 0.5]

fare = np.concatenate([fare, fare_outliers])
distance = np.concatenate([distance, distance_outliers])

data = pd.DataFrame({
    'fare_amount': fare,
    'trip_distance': distance
})

# --- Step 2: Apply Isolation Forest ---
model = IsolationForest(contamination=0.02, random_state=42)
data['anomaly'] = model.fit_predict(data)

# Convert labels (-1 = anomaly, 1 = normal)
data['anomaly'] = data['anomaly'].map({1: 0, -1: 1})

# --- Step 3: Plot results ---
plt.figure(figsize=(8,6))

# Normal points
plt.scatter(data[data['anomaly']==0]['trip_distance'],
            data[data['anomaly']==0]['fare_amount'],
            label='Normal')

# Anomalies
plt.scatter(data[data['anomaly']==1]['trip_distance'],
            data[data['anomaly']==1]['fare_amount'],
            label='Anomaly')

plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.title('Isolation Forest - Anomaly Detection')
plt.legend()
plt.show()
```

Q8.  Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
data = get_rdataset("AirPassengers").data

# Convert to datetime
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)
data.rename(columns={'value': 'Passengers'}, inplace=True)

# --- Train SARIMA Model ---
# (p,d,q) = (1,1,1) and (P,D,Q,s) = (1,1,1,12)
model = SARIMAX(data['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

results = model.fit()

# --- Forecast next 12 months ---
forecast = results.get_forecast(steps=12)
forecast_values = forecast.predicted_mean

# --- Plot results ---
plt.figure(figsize=(10,5))

# Original data
plt.plot(data['Passengers'], label='Original Data')

# Forecast
plt.plot(forecast_values, label='Forecast (Next 12 Months)')

plt.title('SARIMA Forecast - AirPassengers')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.legend()
plt.show()
```

Q9.  Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# --- Step 1: Create sample numerical dataset ---
np.random.seed(42)

# Normal data
X1 = np.random.normal(50, 5, 200)
X2 = np.random.normal(30, 3, 200)

# Add anomalies
X1_outliers = [90, 95, 10]
X2_outliers = [70, 80, 5]

X1 = np.concatenate([X1, X1_outliers])
X2 = np.concatenate([X2, X2_outliers])

data = pd.DataFrame({
    'Feature1': X1,
    'Feature2': X2
})

# --- Step 2: Apply LOF ---
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.02)
labels = lof.fit_predict(data)

# Convert labels (-1 = anomaly, 1 = normal)
data['anomaly'] = np.where(labels == -1, 1, 0)

# --- Step 3: Plot results ---
plt.figure(figsize=(8,6))

# Normal points
plt.scatter(data[data['anomaly']==0]['Feature1'],
            data[data['anomaly']==0]['Feature2'],
            label='Normal')

# Anomalies
plt.scatter(data[data['anomaly']==1]['Feature1'],
            data[data['anomaly']==1]['Feature2'],
            label='Anomaly')

plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('LOF - Anomaly Detection')
plt.legend()
plt.show()
```

Q10. You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.

Explain your real-time data science workflow:

● How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?

● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?

● How would you validate and monitor the performance over time?

● How would this solution help business decisions or operations?


```python
# Install if needed:
# !pip install statsmodels scikit-learn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# -------------------------------
# STEP 1: Create Time Series Data
# -------------------------------
np.random.seed(0)

time = pd.date_range(start='2023-01-01', periods=300, freq='15min')

# Create smooth energy pattern
energy = 50 + 10*np.sin(np.linspace(0, 20, 300)) + np.random.normal(0, 2, 300)

# Add anomalies
energy[50] = 90
energy[200] = 10

df = pd.DataFrame({'energy': energy}, index=time)

# -------------------------------
# STEP 2: Anomaly Detection
# -------------------------------
model = IsolationForest(contamination=0.02, random_state=0)
df['anomaly'] = model.fit_predict(df[['energy']])

# Convert: -1 = anomaly
df['anomaly'] = df['anomaly'].replace({1:0, -1:1})

# Plot anomalies
plt.figure()
plt.plot(df.index, df['energy'])
plt.scatter(df.index[df['anomaly']==1],
            df['energy'][df['anomaly']==1])
plt.title("Anomaly Detection")
plt.show()

# -------------------------------
# STEP 3: Train SARIMA Model
# -------------------------------
train = df['energy'][:200]
test = df['energy'][200:]

model = SARIMAX(train,
                order=(1,1,1),
                seasonal_order=(1,1,1,96),
                enforce_stationarity=False,
                enforce_invertibility=False)

result = model.fit(disp=False)

# -------------------------------
# STEP 4: Forecast
# -------------------------------
forecast = result.forecast(steps=len(test))

# -------------------------------
# STEP 5: Accuracy
# -------------------------------
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

print("MAE:", mae)
print("RMSE:", rmse)

# -------------------------------
# STEP 6: Plot Forecast
# -------------------------------
plt.figure()
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, forecast, label="Forecast")
plt.legend()
plt.title("SARIMA Forecast")
plt.show()
```

In a power grid monitoring system where data is collected every 15 minutes, a real-time data science workflow involves both anomaly detection and short-term forecasting to ensure efficient operations. For detecting anomalies in streaming data, I would use the Isolation Forest algorithm because it is fast, scalable, and well-suited for real-time applications. It works by isolating unusual data points, such as sudden spikes or drops in energy consumption, which may indicate faults, outages, or power theft. Compared to LOF and DBSCAN, Isolation Forest is more practical for continuous data streams as it requires less computational effort and handles large datasets effectively.

For short-term forecasting, I would use the SARIMAX model, as it can capture not only trend and seasonality (such as daily or weekly energy usage patterns) but also incorporate external factors like weather conditions and region. This makes it more accurate than ARIMA and SARIMA in real-world scenarios where energy demand is influenced by multiple variables.

To validate and monitor performance over time, I would use time-based train-test splitting and evaluate the model using metrics such as MAE, RMSE, and MAPE. In a real-time setting, I would continuously monitor prediction errors and anomaly detection rates to identify model drift or performance degradation. If the error increases beyond a certain threshold, the model would be retrained using recent data to maintain accuracy.

This solution provides significant business value by enabling accurate energy demand forecasting, which helps in efficient power generation and distribution planning. Real-time anomaly detection allows quick identification of issues like equipment failure or abnormal consumption patterns, reducing downtime and operational risks. Overall, it improves grid reliability, optimizes resource usage, and supports better decision-making in energy management.
